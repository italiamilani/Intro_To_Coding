```python
import random
#setting up the player class
class Player():
    def __init__(self, name):
        #setting up initial player where they start with 100 health and 0 wins
        self.name = name
        self.health = 100
        self.wins = 0

    def damage(self, amount):
        #creating a function to subtract damage from the players health
        self.health -= amount 
        if self.health < 0:
            self.health = 0
    
    def display_health(self):
        #creating a function to display players current health
        print(esc('34;1') + f'Your health is now {self.health}.' + esc(0))
    
    def win(self):
        self.wins += 1
        
def losingmessage(self):
    print("")
    print(esc('33;1') + "You lose :(" + esc(0))
    
def winningmessage(self):
    print("")
    print(esc('36;1') + "CONGRATULATIONS! You win!" + esc(0))
    self.win()
        
def esc(code):
    return f'\033[{code}m'

#setting up the first room
def room1(self):
    print("-------------------------------------------------------------")
    self.health = 100 #resetting the health for when players play again
    print("You wake up in a mysterious room and instinctively scan your surroundings. " +
          "A door initially draws your attention indicating a way out. " +
          "But a figure in the shadows catches your eye. " +
          "Upon closer examination, you discover it's a minion carrying a cheese ray. " +
          "The minion appears preoccupied with the ray, unaware of your awakening.")
    engage_room1 = input("Do you want to 1) engage with the minion or 2) attempt to slip by? ")
    
    while engage_room1 not in ["1", "2"]: #the input will loop until players input 1 or 2
        print("Please only input 1 or 2.")
        engage_room1 = input("Do you want to 1) engage with the minion or 2) attempt to slip by? ")
    
    if engage_room1 == "1": 
        print(" ")
        print("Choosing the element of surprise, you launch an attack on the unsuspecting minion.")
        print("You manage to defeat the minion and come out unharmed. " +
              "Taking advantage of the situation, you run to the door.")
        room2(self)
    
    if engage_room1 == "2":
        print(" ")
        minion1_sight = random.randint(1,2)
        if minion1_sight == 1: #if the random integer is 1, the minion doesnt see you
            print("You carefully avoid the minion's attention as you quietly move along the wall toward the door.") 
            room2(self)
            
        else: #if not, the minions still see you
            room1_damage = random.randint(1, 20)
            print("You attempt to slide along the wall, but your efforts fall short as the minion sees you and swiftly fires its cheese ray " + esc('31;1') + f'dealing {room1_damage} damage' + esc(0) + " on you.")
            self.damage(room1_damage)
            self.display_health()
            if self.health > 0:
                print("You quickly recover to attack, and you defeat the minion. You take a moment to catch your breath before heading towards the door again.")
                room2(self) #calling the next room
            else:
                print("You try to retaliate, but your legs have turned into cheese. The minion gets closer before firing one last time.")
                losingmessage(self)
        
def room2(self):
    print(esc('37') + "-------------------------------------------------------------" + esc(0))
    print("When you open the second door, you notice 2 minions wielding cheese rays and another door.")
    engage_room2 = input("Do you want to 1) engage with the minions or 2) try to sneak by? ")

    while engage_room2 not in ["1", "2"]:
        print("Please only input 1 or 2.")
        engage_room2 = input("Do you want to 1) engage with the minions or 2) try to sneak by? ")
    
    if engage_room2 == "1":
        print(" ")
        room2_damage = random.randint(10, 30)
        print("Taking the first minion by surprise, you overpower it swiftly.")
        print("However, the second one quickly responds, firing its cheese ray and " + 
              esc('31;1') + f'deals {room2_damage} damage' + esc(0) + " to you.")
        self.damage(room2_damage)
        self.display_health()
        if self.health > 0:
            print("Undeterred, you retaliate and swiftly defeat the second minion. " +
              "You take a moment to catch your breath before heading towards the next door.")
            room3(self)
        else:
            print("You try to retaliate, but your legs have turned into cheese. The minion gets closer before firing one last time.")
            losingmessage(self)

    if engage_room2 == "2":
        print("")
        minion2_sight = random.randint(1,3)
        if minion2_sight == 1:
            print("You carefully avoid the minions' attention as you quietly move along the wall toward the door.")
            room3(self)
        else:
            room2_damage1 = random.randint(10, 30)
            print("You attempt to slide along the wall, but your efforts fall short as one of the minions spots you and swiftly fires its cheese ray, " + 
                  esc('31;1') + f'dealing {room2_damage1} damage' + esc(0) + " on you.")
            self.damage(room2_damage1)
            self.display_health()
            if self.health > 0:
                room2_damage2 = random.randint(10, 30)
                print("You quickly recover to attack, and you defeat the first minion.")
                print("The second minion fires their cheese gun, " + 
                      esc('31;1') + f'dealing {room2_damage2} damage' + esc(0) + " to you.")
                self.damage(room2_damage2)
                self.display_health()
                if self.health > 0:
                    print("You strike back and manage to defeat the second minion. " + 
                          "You take a moment to catch your breath before heading towards the door again.")
                    room3(self)
                else:
                    print("You try to retaliate, but your legs have turned into cheese. The minion gets closer before firing one last time.")
                    losingmessage(self)
            else:
                print("You try to retaliate, but your legs have turned into cheese. The minion gets closer before firing one last time.")
                losingmessage(self)

def room3(self):
    print(esc('37') + "-------------------------------------------------------------" + esc(0))
    print("When you open the third door, you notice 3 minions wielding cheese rays and another door.")
    engage_room3 = input("Do you want to 1) engage with the minions or 2) try to sneak by? ")
    
    while engage_room3 not in ["1", "2"]:
        print("Please only input 1 or 2.")
        engage_room3 = input("Do you want to 1) engage with the minions or 2) try to sneak by? ")
    
    if engage_room3 == "1":
        print(" ")
        room3_damage1 = random.randint(10, 30)
        print("You manage to take the first minion by surprise, overpowering him.")
        print("But the second one quickly responds, firing and " +
              esc('31;1') + f'dealing {room3_damage1} damage' + esc(0) + " to you.")
        self.damage(room3_damage1)
        self.display_health()
        if self.health > 0:
            room3_damage2 = random.randint(10, 30)
            print("You swiftly retaliate against the second minion, overpowering him as well.")
            print("But the third one quickly responds, firing and " +
                  esc('31;1') + f'dealing {room3_damage2} damage' + esc(0) + " to you.")
            self.damage(room3_damage2)
            self.display_health()
            if self.health > 0:
                print("You swiftly retaliate against the third minion, overpowering him as well. " +
                      "You take a moment to catch your breath before heading towards the next door.")
                boss(self)
            else:
                print("You try to retaliate, but your legs have turned into cheese. The minion gets closer before firing one last time.")
                losingmessage(self)
        else:
            print("You try to retaliate, but your legs have turned into cheese. The minion gets closer before firing one last time.")
            losingmessage(self)
            
    if engage_room3 == "2":
        print("")
        minion3_sight = random.randint(1,4)
        if minion3_sight == 1:
            print("You carefully avoid the minions attention as you quietly move along the wall toward the door.")
            boss(self)
        else:
            room3_damage1 = random.randint(10, 30)
            print("You attempt to slide along the wall, but your efforts fall short as one of the minions spots you " +
                 "and swiftly fires its cheese ray " +
                 esc('31;1') + f'dealing {room3_damage1} damage' + esc(0) + " to you.")
            self.damage(room3_damage1)
            self.display_health()
            if self.health > 0:
                print("You quickly recover to attack, and you defeat the first minion.")
                room3_damage2 = random.randint(10, 30)
                print("The second minion fires their cheese gun, " + 
                      esc('31;1') + f'dealing {room3_damage2} damage' + esc(0) + " to you.")
                self.damage(room3_damage2)
                self.display_health()
                if self.health > 0:
                    print("You strike back and manage to defeat the second minion.")
                    room3_damage3 = random.randint(10, 30)
                    print("The third minion fires their cheese gun, " + 
                          esc('31;1') + f'dealing {room3_damage3} damage' + esc(0) + " to you.")
                    self.damage(room3_damage3)
                    self.display_health()
                    if self.health > 0:
                        print("You strike back and defeat the third minion. " +
                              "You take a moment to catch your breath before heading towards the next door.")
                        boss(self)
                    else:
                        print("You try to retaliate, but your legs have turned into cheese. The minion gets closer before firing one last time.")
                        losingmessage(self)
                else:
                    print("You try to retaliate, but your legs have turned into cheese. The minion gets closer before firing one last time.")
                    losingmessage(self)
            else:
                print("You try to retaliate, but your legs have turned into cheese. The minion gets closer before firing one last time.")
                losingmessage(self)
    
def boss(self):
    boss_health = 100
    print(esc('37') + "-------------------------------------------------------------" + esc(0))
    print("You see a large figure facing a window with their back to you as you cautiously open the door. " +
         "A table covered with a variety of weapons to your left catches your attention. " +
         "You sneak inside and inspect the weapons on the table.")
    boss_weapon = input("Do you want to pick up the " + 
                        esc('32;1') + "1) Fart Gun" + esc(0) + " or " + 
                        esc('35;1') + "2) Lipstick Taser" + ": ") #choose between 2 weapons that Gru normally has
    
    while boss_weapon not in ["1", "2"]:
        print("Please input 1 or 2.")
        boss_weapon = input("Do you want to pick up the " + 
                            esc('32;1') + "1) Fart Gun" + esc(0) + " or " + 
                            esc('35;1') + "2) Lipstick Taser" + ": ")

    if boss_weapon == "1":
        print("")
        print("The figure turns towards you as you pick up " +
              esc('32;1') + "the Fart Gun"  + esc(0) + ", and it's Dru. " +
              "You automatically point and shoot. " +
              "A " + esc('32') + "green cloud"  + esc(0) + 
              " shoots towards Dru as the gun lets out a loud fart noise. " +
              "The smell easily overwhelms Dru, causing him to pass out immediately.")
        print("You have successfully defeated Dru, discovering a large exit sign and escaping his mansion.")
        winningmessage(self)
    
    if boss_weapon == "2":
        print("")
        print("The figure turns towards you as you pick up the Lipstick Taster, and it's Dru.")
        while boss_health > 0 and self.health > 0:
            print("You point and shoot " +
                  esc('31;1') + "dealing 40 damage" + esc(0) + " to Dru.")
            boss_health -= 40
    
            if boss_health <= 0:
                print("At this blow, Dru falls to the ground. Using this to your advantage, you look around to " +
                      "find a large exit sign and run to escape his mansion")
                winningmessage(self)
                break

            print("Dru attacks you back, " +
                  esc('31;1') + "dealing 20 damage" + esc(0) + ".")
            self.damage(20)
            self.display_health()
    
            if self.health <= 0:
                print("Despite your valiant efforts, Dru's attack takes its toll, and you succumb to the damage.")
                losingmessage(self)
                break

def game(self):
    print(esc('0;1') + "Welcome to Defeat Dru! Try to escape his castle while fighting off his evil minions." + esc(0))
    play = input("Do you want to play Defeat Dru? Y/N : ")
    
    while play.lower() not in ["y", "n"]: #the input will loop until players input y or n
        print("Please only input Y or N.")
        play = input("Do you want to play Defeat Dru? Y/N: ")
    
    while play.lower() == "y":
        room1(self)
        play = input("Do you want to play again? Y/N : ")
        while play.lower() not in ["y", "n"]: #the input will loop until players input y or n
            print("Please only input Y or N.")
            play = input("Do you want to play again? Y/N: ")
    
    if self.wins == 1:
        print("Thanks for playing! You have defeated Dru", self.wins, "time.") #printing your total wins each time you play
    elif self.wins >= 2:
        print("Thanks for playing! You have defeated Dru", self.wins, "times.") 
    else:
        print("Thanks for playing! You did not defeat Dru once, maybe next time.")

player1 = Player("Gru") #creating a class player of Gru, Dru's brother and who the player plays as       

game(player1)
```
