# Post Lab

_1. create a clean dataframe with a new variable_

```python
realestate = ct_re.copy()
```

_2. rename all of the column names and take the space out_

```python
realestate.columns = ["Serial", "List_Year", "Date", "Town", "Address", "Assess_Value", "Sale_Amount", "Sales_Ratio", "Property_Type", "Residential_Type", "Code", "Remarks", "OPM_Remarks", "Location"]

realestate
```

_3. isolate the list years between 2001-2009_

```python
realestate.drop(realestate[(realestate["List_Year"] < 2001)].index, inplace = True)
realestate.drop(realestate[(realestate["List_Year"] > 2009)].index, inplace = True)

realestate
```

_4. using a for loop and dictionary, count the number of houses listed in each town_

```python
town = realestate["Town"]
town_dict = {}

for i in town:
    if i not in town_dict:
        town_dict[i] = 1
    else:
        town_dict[i] += 1

town_dict
```

_5. find the mean of the "assessed value" column_

```python
realestate['Assess_Value'].mean()
```

_6. remove all the numbers from the address column_

```python
realestate["Address"] = realestate["Address"].str.replace("\d+", "", regex = True)

realestate
```

_7. create a heatmap to find all the NaN values in the dataset_

```python
import seaborn as sns
```

```python
cols = realestate.columns[:30]
colors = ['#D6F6DD', '#ACECF7']

sns.heatmap(realestate[cols].isnull(), cmap = sns.color_palette(colors))
```

_8. create a copy of your data frame and delete all the rows with NaN values in them_

```python
realestate_2 = realestate.copy()
realestate_2 = realestate_2.dropna()

realestate_2
```

_9. use the original dataframe fill in those NaN values_

```python
realestate = realestate.fillna(0)

realestate
```
